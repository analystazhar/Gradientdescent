# GRADIENT DESCENT ALGORITHM

loading  the data set using sklearn  

```python
from sklearn.datasets import make_regression
```

```python
import numpy as np
```

```python
x,y=make_regression(n_samples=4,n_features=1,n_informative=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

# APPLY LINEAR REGRESSION FIRST

```python
from sklearn.linear_model import LinearRegression
```

```python
reg= LinearRegression()
```

```python
reg.fit(x,y)
```

```python
#finding slope
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')
```

# LETS APPLY GRADIENT DESCENT ASSUMING SLOPE IS CONSTANT=78.35


AND LET'S ASSUME THE STARTING VALUES FOR INTERCEPT B=0

# ITERATION 1


```python
y_pred=((78.35*x + 0).reshape(4))
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='ols')
plt.plot(x,y_pred,color='green',label='b = 0')
plt.legend()
plt.show()
```

```python
m=78.35
b=0
loss_slope=-2 * np.sum(y-m*x.ravel() - b)
```

```python
loss_slope
```

```python
#lets take learning rate
lr=0.1
step_size=loss_slope*lr
step_size
```

```python
#calculating the new intercept
b=b-step_size
b
```

```python
y_pred1=((78.35 * x) + b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='ols')
plt.plot(x,y_pred1,color='blue',label='b ={}'.format(b))
plt.plot(x,y_pred,color='green',label='b = 0')

plt.legend()
plt.show()
y_pred1
```

```python
b=-100
m=78.35
lr=0.1
epochs=100 #number of iteration
for i in range(epochs):
    loss_slope=-2* np.sum(y-m*x.ravel() -b)
    b=b-(lr*loss_slope)
    y_pred=m*x + b
    plt.plot(x,y_pred)
plt.scatter(x,y)
 
```
